# Class that does processing of single article

```python
class NewsArticle():
    # vocabulary was previously a property
    def vocabulary(self, force_compute=False):
        if(force_compute==True):
            if "alpha" in globals() and "beta" in globals():
                self.update_vocab(alpha, beta)
                return self._vocabulary
            else:
                raise Exception("Please define alpha and beta in global scope")
        try:
            v=self._vocabulary
        except AttributeError:
            if "alpha" in globals() and "beta" in globals():
                self.update_vocab(alpha, beta)
                v=self._vocabulary
            else:
                raise Exception("Please define alpha and beta in global scope")
        return v
    
    @property
    def groupA(self):
        return NewsArticle.statistics(self.named_ents + self.title_ents)

    @property
    def groupB(self):
        return NewsArticle.statistics(self.other_noun_and_verb_ents)
    
    #frequency statistics
    @staticmethod
    def statistics(tup_list):
        counts = collections.Counter(tup_list)
        count_tupels=[(x, counts[x]) for x in counts]
        return sorted(count_tupels, key=lambda x:x[1], reverse=True)
    
    @staticmethod
    def POS(text):
        text = nltk.word_tokenize(text)
        text = nltk.pos_tag(text)
        return text
    
    @staticmethod
    def NER(text): 
        article = nlp(text)
        labels = [x.label_ for x in article.ents]
        items = [(x.text, x.label_) for x in article.ents]
        return items
    
    @staticmethod
    # gives the article and removes unwanted html stuff
    def _get_article(url):
        res = requests.get(url, timeout=timeout)
        res.raise_for_status()
        html = res.text
        soup = BeautifulSoup(html, 'html.parser')
        for script in soup(["script", "style", 'aside']):
            script.extract()
        return soup.get_text()
    
    @staticmethod
    # remove punctuations from article
    def _remove_punctuations(text):
        art = " ".join(re.split(r'[\n\t]+', text))
        # remove punctiuations
        punctuations=list(string.punctuation) + [u'\xa0',u'”',u'“',u'•','  ','’','\r'] 
        punctuations.remove("'")
        for p in punctuations:
            art=art.replace(p, "")
        
        return ' '.join(art.split())
    
    @staticmethod
    def _remove_long_words(text):
        threshold=45
        long_words = [word for word in word_tokenize(text) if len(word)>threshold]
        for word in long_words:
            text = text.replace(long_words, "")
        return ' '.join(text.split())
        
    @staticmethod
    # remove stop words
    def _remove_stop_words(text):
        stop_words = set(stopwords.words('english')) 
        word_tokens = word_tokenize(text) 
        filtered_sentence = [w for w in word_tokens if not w in stop_words] 
        return ' '.join(filtered_sentence) 
    
    @staticmethod
    # remove b from a
    def _remove_ner(article, ner_t):
        for x,y in ner_t:
            article = article.replace(x, "")
        return ' '.join(article.split())
    
    # prepare all entitiy tables
    def _prepare_ents(self):
        self.prepared_text=NewsArticle._remove_punctuations(self.text)
        self.prepared_text=NewsArticle._remove_punctuations(self.prepared_text)
        
        #time and named entities
        temp_ner = NewsArticle.NER(self.prepared_text)
        self.time_ents= [(tup[0],tup[1]) for tup in temp_ner if tup[1] in ['DATE'] ]
        self.named_ents = [(tup[0],tup[1]) for tup in temp_ner if tup[1] not in ['DATE'] ]
        
        # title entities
        not_useful_ents=["CC","DT","EX","IN","PDT","POS","PRP","PRP$","TO","UH","WDT","WP","WP$"]
        temp_title_pos = NewsArticle.POS(self.title.lower())
        self.title_ents = [tup for tup in temp_title_pos if tup[1] not in not_useful_ents] #Not of Not useful is useful
        
        # other entities from the article
        self.other_text = NewsArticle._remove_ner(NewsArticle._remove_stop_words(self.prepared_text.lower()), temp_ner) # remove already counted named,time entities
        temp_other_pos = NewsArticle.POS(self.other_text)
        noun_and_verb_labels = ["NN","NNS","NNP","NNPS","VB","VBD","VBG","VBN","VBP","VBZ"] #only nouns and verbs are required
        self.other_noun_and_verb_ents = [(tup[0],tup[1]) for tup in temp_other_pos if tup[1] in noun_and_verb_labels]
        
    def __init__(self, url, title='', published_time=0, event_id=''):
        self.url=url
        self.title=title
        self._event_id=event_id
        try:
            self.published_time=datetime.fromtimestamp(published_time)
        except:
            self.published_time=datetime.fromtimestamp(0)
        self.text=NewsArticle._get_article(url)
        
        #prepare entities
        self._prepare_ents()

    # print the article 
    def __str__(self):
        return "Time ents\n\n" + pformat(self.time_ents) + \
               "\n\nNamed ents\n\n" + pformat(self.named_ents) + \
               "\n\nOther ents\n\n" + pformat(self.other_noun_and_verb_ents) \
    
    #print the prepared article
    def __repr__(self):
        return self.__str__()
    
    def update_vocab(self, alpha, beta):
        vocabulary_temp = []
        N = len( self.time_ents + self.named_ents + self.other_noun_and_verb_ents + self.title_ents)
        
        a_words_len = int(alpha*beta*N)
        a_words = self.groupA[:a_words_len]
        
        b_words_len = int(alpha*N) - len(a_words)
        b_words = self.groupB[:b_words_len]
        
        vocabulary_temp.extend(a_words)
        vocabulary_temp.extend(b_words)
        
        self._vocabulary=dict({word[0][0]:word[1] for word in vocabulary_temp})
        # remove
        #print("update_vocab", vocabulary_temp)
   
```

# Create overall vocabulary for TF-IDF

```python
# function for creating overall vocab
def overall_vocab(articles, force_compute=False):
    vocab = dict()
    indices = dict()
    i=0
    for article in articles:
        for word in article.vocabulary(force_compute).keys():
            if word in vocab:
                vocab[word]+=1
            else:
                vocab[word]=1
                indices[word]=i
                i+=1
    return (vocab,indices)
vocab, indices = overall_vocab(articles, True)
```

# Create document vectors based on TF-IDF for all articles

```python
# function for creating document vector of an article
def vector(article,vocab,indices,N):
    v = [0 for i in range(len(indices))]
    for word in article.vocabulary().keys():
        tf = article.vocabulary()[word]
        idf = np.log10(N/vocab[word] + 0.01 )
        w = tf*idf
        v[indices[word]] = w 
    v = np.array(v)
    val = np.sqrt(np.sum(v**2))
    return (v/val)[:]

for article in articles :
    article.docvector = vector(article, vocab=vocab, indices=indices, N=len(articles))
    #print(article.docvector, end="\n\n")
    #print(np.sum(article.docvector**2), end="\n\n\n\n") # it will be around 1
```

# Define the similarity function

```python
def similarity(v1,v2):
    assert len(v1) == len(v2)
    return 1000*min(np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2)),1.0)
```

# Define the Event class: an event is collection of articles

```python
class Event:
    def vocabulary(self, dummy_force=False):
        return self._vocabulary
        
    def __init__(self):
        self.article_count = 0
        self.articles = []
        self.docvector = []
        self._vocabulary = {}
    
    # add updates the docvector automaticlally
    def add(self, article, similarity):
        assert(isinstance(article, NewsArticle))
        article.similarity = similarity
        self.articles.append(article)
        self.article_count+=1
        article_vocab = article.vocabulary()
        for i in article_vocab:
            if i in self._vocabulary:
                self._vocabulary[i]+=article_vocab[i]
            else:
                self._vocabulary[i]=article_vocab[i]
        self._article_id_count = Counter([article._event_id for article in self.articles])
        self._event_id = max(self._article_id_count, key=lambda x: self._article_id_count[x])
    
    def update_vector(self, vocab, indices, N):
        self.docvector = vector(self,vocab,indices,N)[:]

    def __str__(self):
        return "Event %s, Total articles %s\n" %(self._event_id, self.article_count) +\
                ('\n'.join(str((article.title[:30]+'..', article._event_id[:10]+'..', article.similarity)) for article in sorted(self.articles,key=lambda x:x.similarity,reverse=True)) if self.article_count!=0 else "No article found") + \
                '\n' + str(self._article_id_count)
    
    def __repr__(self):
        return self.__str__()

def metrics(articles, events, CE=CE, CF=CF):
    # articles related to a topic
    actual_event_articles = {}
    for article in articles:
        if article._event_id in actual_event_articles:
            actual_event_articles[article._event_id].append(article)
        else:
            actual_event_articles[article._event_id]=[article]
    overall_cost = 0
    for event in events:    
        A = len(actual_event_articles[event._event_id])
        A1 = event._article_id_count[event._event_id]
        B = len(articles) - A
        B1 = event.article_count - A1
        PE = B1/B
        PF = (A-A1)/A
        #print("false positive, true negetive = ", PE,PF)
        P= A/len(articles)
        overall_cost += (CE*PE*P + CF*PF*(1-P))**2
    return math.sqrt(overall_cost/len(events))
```

# Function to cluster articles based on the single pass algo described in paper

```python
# function to cluster articles into events and return events
def cluster_articles(articles, alpha_l, beta_l, threshold):
    vocab, indices = overall_vocab(articles, True)
    global alpha
    global beta
    alpha = alpha_l
    beta = beta_l
    for article in articles :
        article.docvector = vector(article, vocab=vocab, indices=indices, N=len(articles))
    events = []
    for article in articles:
        # event, similarity, e_index
        #print(events,end='\n\n\n')
        potential = (None,-1,-1)
        for i,event in enumerate(events):
            cur_sim = similarity(event.docvector, article.docvector)
            if potential[1]<cur_sim: 
                potential = (event,cur_sim,i)
        if(potential[1]>=threshold):
            #print("adding article: %s to event"% article.title,potential[2])
            potential[0].add(article,potential[1])
            potential[0].update_vector(vocab, indices, len(articles))
        else:
            #print("Creating New Event for article:", article.title)
            e = Event()
            e.add(article,similarity(article.docvector, article.docvector))
            e.update_vector(vocab, indices, len(articles))
            events.append(e)
        article._event_assigned=potential[2]
    return events
```

# Finding optimal alpha and beta values

```python
%%time

# set actually calculate to true for calculating all this stuff (takes long time:: 30-60 mins)
# set to false to load from dumped json file
actually_calculate = False

if(actually_calculate == True or not os.path.exists('acc_list.json')):
    #create a event for first article
    acc_list = []
    #for th in list(range(0,1000,5))+ [20]:
    for al in range(0,101,5):
        alpha = al/100
        for be in range(0,101,5):
            # because i am changing alpha and beta
            beta = be/100
            vocab, indices = overall_vocab(articles, force_compute=True)
            events = cluster_articles(articles, alpha, beta, threshold)
            acc_list.append((alpha, beta, metrics(articles, events), len(events)))
        print("running for alpha=", alpha)
    json.dump(acc_list, open("acc_list.json", "w"), indent=2)
else:
    try:
        acc_list1 = json.load(open("acc_list.json", "r"))
        acc_list=acc_list1[:]
    except:
        acc_list = []
        raise Exception("File Not found, set actually_calculate to true")

print()        
pprint(sorted(acc_list, key = lambda x: x[2]))
```

# Run clustering for best alpha and beta values (on training data)

```python
%%time
best_tuple = min(acc_list, key = lambda x: x[2])
best_alpha, best_beta = best_tuple[0], best_tuple[1]
print("best alpha=%s, best beta=%s"%(best_alpha, best_beta))

# run for best tuples
events = cluster_articles(articles, best_alpha, best_beta, threshold)
print("cost=", metrics(articles, events), end='\n\n')
for event in events:
    pprint(event)
    print()
```
